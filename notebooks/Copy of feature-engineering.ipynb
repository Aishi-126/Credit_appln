```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('credit-score.csv')
df.head()
```

```python

loan_types = []
for index in df.index:
    loan = df.Type_of_Loan[index]
    if isinstance(loan, str):
        temp = loan.replace('and ', '').split(', ')
        for i in temp:
            if i not in loan_types:
                loan_types.append(i)
```

```python

for loan in loan_types:
    df[loan.replace(' ', '_')] = 0  # create the loan column in the df with 0
    for index in df.index:
        loan_value = df.Type_of_Loan[index]
        if isinstance(loan_value, str):
            temp = loan_value.replace('and ', '').split(', ')
            if loan in temp:
                df.loc[index, loan.replace(' ', '_')] = 1
```

```python
df.iloc[:,-9:].head()
```

```python

month = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8
}
df.Month.replace(month, inplace=True)
df.Month = df.Month.astype(int)
```

```python
print(df['Credit_Mix'].unique())
```

```python
import numpy as np

# Define the mapping dictionary
credit_mix = {
    'Standard': 0,
    'Good': 1,
    'Bad': -1
}

# Replace values according to the mapping dictionary
df['Credit_Mix'] = df['Credit_Mix'].replace(credit_mix)

# Convert any non-numeric values to NaN
df['Credit_Mix'] = pd.to_numeric(df['Credit_Mix'], errors='coerce')

# Convert the column to integers
df['Credit_Mix'] = df['Credit_Mix'].astype('Int64')  # Using 'Int64' to support NaN values


```

```python

credit_score = {
    'Standard': 0,
    'Poor': -1,
    'Good': 1
}
df.Credit_Score.replace(credit_score, inplace=True)
df.Credit_Score = df.Credit_Score.astype(int)
```

```python
df['Missed_Payment_Day'] = 0
for row in df.index:
    if df.loc[row, 'Delay_from_due_date'] > 0:
        df.loc[row, 'Missed_Payment_Day'] = 1
```

```python
df.to_csv('credit-score.csv', index=False)
```
